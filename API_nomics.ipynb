```python
import pandas as pd
from pandas import DataFrame
```

```python
from nomics import Nomics
nomics = Nomics("7aa538ba6a64a099bbbb9add78412550")

markets = nomics.Markets.get_markets(exchange = 'binance')
print(markets)
```

```python
WWW = markets
pd.set_option('display.max_rows', None)
rrr = DataFrame (WWW,columns=['exchange','market','base','quote'])
print(rrr.head(10))
```

```python
nomics = Nomics("7aa538ba6a64a099bbbb9add78412550")

data = nomics.Volume.get_volume_history()

print(data)

Volume = data
pd.set_option('display.max_rows', None)
PPP = DataFrame (Volume,columns=['timestamp','volume',])
print(PPP.tail(10))
```
